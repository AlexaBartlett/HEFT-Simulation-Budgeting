```python
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, "/Users/alexabartlett/class/python/build/lib.macosx-10.9-x86_64-3.9")
from classy import Class
import pyccl as ccl
import pyccl.nl_pt as pt
import numpy as np
import scipy as sc
```

```python
k = np.loadtxt('training_data/pmm_w0wamnu_noisy_ZCV_tier1/k_out.txt')[:539]
default_100 = np.loadtxt('emulator_errors/example_kerr.txt', unpack=False)
npc8_100 = np.loadtxt('emulator_errors/example_npc8_kerr.txt', unpack=False)
npc12_100 = np.loadtxt('emulator_errors/example_npc12_kerr.txt', unpack=False)
ord2_100 = np.loadtxt('emulator_errors/example_order2_kerr.txt', unpack=False)
ord4_100 = np.loadtxt('emulator_errors/example_order4_kerr.txt', unpack=False)

default_1000 = np.loadtxt('emulator_errors/tier2_XL_default_kerr.txt', unpack=False)
npc8_1000 = np.loadtxt('emulator_errors/tier2_XL_npc8_kerr.txt', unpack=False)
npc12_1000 = np.loadtxt('emulator_errors/tier2_XL_npc12_kerr.txt', unpack=False)
ord2_1000 = np.loadtxt('emulator_errors/tier2_XL_ord2_kerr.txt', unpack=False)
ord4_1000 = np.loadtxt('emulator_errors/tier2_XL_ord4_kerr.txt', unpack=False)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4), sharey='row')

axs[0].semilogx(k, npc8_100[1], label=r'$N_{\rm PC}=8$')
axs[0].semilogx(k, npc12_100[1], label=r'$N_{\rm PC}=12$')
axs[0].semilogx(k, ord2_100[1])
axs[0].semilogx(k, ord4_100[1])
axs[0].semilogx(k, default_100[1], c='k', ls='--')
axs[0].legend(title=r'$N=100$', ncol=2, fontsize=12, title_fontsize=12)

axs[1].semilogx(k, npc8_1000[1])
axs[1].semilogx(k, npc12_1000[1])
axs[1].semilogx(k, ord2_1000[1], label=r'Order$=2$')
axs[1].semilogx(k, ord4_1000[1], label=r'Order$=4$')
axs[1].semilogx(k, default_1000[1], label='Default', c='k', ls='--')
axs[1].legend(title=r'$N=1000$', ncol=2, fontsize=12, title_fontsize=12)

fig.supxlabel(r'$k\quad[h\,{\rm Mpc}^{-1}]$')
axs[0].set_ylabel('Fractional Emulator Error', fontsize=12)
fig.tight_layout()
fig.show()

plt.savefig('figures/hyperparams.pdf')
```
